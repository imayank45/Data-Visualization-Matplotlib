### What is matplotlib?
Matplotlib is a popular open-source Python library used for creating static, animated, and interactive visualizations in a wide range of fields, including data analysis, scientific research, and data science.

### Importing Libraries

```python
import numpy as np
import matplotlib as mp
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.datasets import load_iris
iris = load_iris()
```

### 1. Generating sin and cos graph using matplotlib

```python
x = np.linspace(0, 4*np.pi, 100)

# defining wave
sin_y = np.sin(x)
cos_y = np.cos(x)

# plotting the wave
plt.plot(x, sin_y, label = 'Sine')
plt.plot(x, cos_y, label = 'Cosine')

plt.title('Sine and Cosine Waves')

plt.legend(loc = 'upper right')
plt.show()
```

```python
x = np.linspace(0, 4*np.pi, 100)

# defining wave
sin_y = np.sin(x)
cos_y = np.cos(x)

# plotting the wave
plt.plot(x, sin_y,  '*')
plt.plot(x, cos_y,  '--')

plt.title('Sine and Cosine Waves')

plt.legend(loc = 'upper right')
plt.show()
```

### 2. Setting styles to the graph

```python
plt.figure(figsize = (15,5))

plt.style.use('grayscale')

plt.plot(x, sin_y, label = 'Sine')
plt.plot(x, cos_y, label = 'Cosine')

plt.title('Sine and Cosine Waves')

plt.legend(loc = 'upper right')
plt.show()
```

### 3. Saving the plots in our local machine

```python

# Generate some data
x = np.linspace(0, 4 * np.pi, 100)
sin_y = np.sin(x)
cos_y = np.cos(x)

# Create a figure
plt.figure(figsize=(15, 5))
plt.style.use('seaborn-dark')

# Plot the sine and cosine curves
plt.plot(x, sin_y, label='Sine')
plt.plot(x, cos_y, label='Cosine')

# Add a title
plt.title('Sine and Cosine Waves')

# Save the figure as an image
plt.savefig('Sine & Cosine.png')

# Display the figure
plt.show()
```

### 4. Different types of plots

#### 1. Line plot

```python
x = np.linspace(0, 4*np.pi, 100)
y = np.sin(x)

plt.style.use('Solarize_Light2')

plt.plot(x,y)

plt.title('Sine and Cosine Waves')

plt.show()
```

#### 2. Multiple line plots

```python
x = np.linspace(0, 4 * np.pi, 100)

y_sin = np.sin(x)
y_cos = np.cos(x)

plt.figure(figsize = (15,5))
plt.style.use('Solarize_Light2')

plt.plot(x, y_sin, '*', color = 'red', label = 'Sine')
plt.plot(x, y_cos, '--', color = 'blue', label = 'Cosine')

plt.title('Sine and Cosine Wave')
plt.legend()
plt.show()
```

### Multiple Subplots
Multiple subplots refers to dividing a figure into multiple sub-plots or axis, each of which can contain a seperate plot. This allows for multiple plots to be displayed in single figure. In matplotlib you can ceate multiple subplots using subplot function.

```python
x = np.linspace(1,10,50)
fig, ax = plt.subplots(2,2)

y = np.sin(x)
ax[0,0].plot(x, y)
ax[0,0].set_title('Sin Wave')

y = np.cos(x)
ax[0,1].plot(x, y)
ax[0,1].set_title('Cos Wave')

y = np.tan(x)
ax[1,0].plot(x, y)
ax[1,0].set_title('Tan Wave')

y = np.random.randn(1000)
ax[1,1].hist(y, bins = 40, edgecolor = 'black')
ax[1,1].set_title('Random Distribution')

plt.tight_layout()
plt.show()
```

```python
plt.subplots_adjust(hspace = 0.4, wspace = 0.4)

for i in range(1, 7):
    
    plt.subplot(2, 3, i)
    plt.text(0.5, 0.5, str((2, 3, i)), fontsize = 18, ha = 'center')
    
```

```python
x = np.linspace(1,10,50)
y = np.sin(x)

for i in range(1,2):
    
    fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize = (18,6))
    
    ax1.plot(x, y)
    ax2.plot(x, y**2)
    ax3.plot(x, y**3)
    
    ax1.set_title('Sin Wave')
    ax2.set_title('Sin Wave - Square')
    ax3.set_title('Sin Wave - Cube')
    
    plt.tight_layout()
    plt.show()
```

### 3D Plots
Matplotlib provides a toolkit for 3D plotting named mplot3d, which provides functionality to create 3D plots.

```python
plt.axes(projection = '3d')
```

```python
ax = plt.axes(projection = '3d')

z = np.linspace(0, 15, 1000)
x = np.sin(z)
y = np.cos(z)

ax.plot3D(x, y, z, c = 'purple')
```

```python
ax = plt.axes(projection = '3d')

z = np.linspace(0, 15, 1000)
y = np.sin(z)
x = np.cos(z)

ax.plot3D(x, y, z, c = 'purple')
```

```python
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure()
ax = fig.add_subplot(111, projection = '3d')

x = np.random.randn(100)
y = np.random.randn(100)
z = np.random.randn(100)


ax.scatter(x, y, z, c = 'red', marker = '*', s = 100)

ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis')

plt.tight_layout()

plt.show()
```
